```python
def factorial(num):
    if num<=1:
        return 1
    else:
        return num*factorial(num-1)
    return
print(factorial(5))
```

## Limitation of Recursion
<ul>
    <li> Python stops the recursion after certain number depths - after 2950 probably (if the program doesn't return anything)</li>
</ul>


```python
print(factorial(3000))
```

### Solution for the above problem<br>
Import system library and set the recursion limit

```python
import sys
sys.setrecursionlimit(5000)
print(factorial(2980))
```

### Fibonacci series

```python
def fibo(n):
    result = []
    a,b = 0,1
    while(a < n):
        result.append(a)
        a,b = b, a+b
    return result
print(fibo(10))
```

### Fibonacci in recursion

```python
def fibo(n):
    result = []
    if(n <= 1):
        return 1
    else:
        return fibo(n-1)+fibo(n-2)
for i in range(10):
    print(fibo(i))
```
